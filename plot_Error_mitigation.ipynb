```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 400
plt.rcParams['figure.figsize'] = (6, 4.5)


# font = {'family' : 'serif',
#         'size'   : 9}

# plt.rc('font', **font)
```

```python
df = pd.read_csv("plot-data.txt")
```

```python
mitigated_E = [-0.542064571489493, -0.8516371488365992, -1.0309501636120274, -1.084253157012312, -1.0650694734588316 , -1.0651732801528115 , 
               -0.9930620135684327, -0.9383694394794386, -0.9283694394794386]
Non_mitigated_E = [-0.5029865640269612, -0.8011371533265942, -0.9819502176821252, -1.024154158314517, -1.0361695832682401, -1.042674332134352 ,
                   -0.9641482728634113, -0.9203085234996417, -0.8983694394794386]
distance = [0.3,0.4,0.6,0.73,0.9,1,1.2,1.4,1.6]
```

```python
#plt.plot(df['Bond-Length'], df[' VQE Energy(electronic)'], '-o')
plt.plot(df['Bond-Length']/10, df[' VQE Energy(molecular)'], '-o', markersize = 5, linewidth= 0.5, label='Simulated')
plt.plot(df['Bond-Length']/10, df[' Exact Energy(molecular)'], '-o', markersize = 5, linewidth= 0.5, label='Exact Diagonalization')
plt.plot(distance, mitigated_E, 'o', markersize = 5, linewidth= 0.5, label='QIBM(Mitigated)')
plt.plot(distance, Non_mitigated_E, 'o', markersize = 5, linewidth= 0.5, label='QIBM(Non-mitigated)')
plt.ylabel(r'$E_{molecular}$' + ' (Hartree)')
plt.xlabel(r'Bond Length ($\AA$)')
plt.legend()
plt.grid(alpha=0.2)
plt.savefig('Error_mitigation-curve.pdf')

plt.show()
```
